```python
import chardet
```

```python
def find_encoding(fname): #이 함수 항상 기억하기.
    r_file = open(fname,'rb').read() #open으로 리딩하는 버퍼링 공간
    result = chardet.detect(r_file)#인코딩 정보 확인.
    charenc = result['encoding']
    return charenc
```

```python
my_encoding1 = find_encoding('nav_2018.csv')

df_1 = pd.read_csv('nav_2018.csv', encoding=my_encoding1)
```

```python
games_all=games_naser.copy()

tidy = pd.melt(games_all.reset_index(),
               id_vars=['게임', '경기일자'], value_vars=['방문팀', '홈팀'],
               var_name='팀구분',value_name='팀')
#대상, 고정열, 합쳐질 열, 합쳐진 열의 이름, 값의 이름
```

```python
tidy
```

```python
tidy.경기일자.dtype #합쳐졌어도 데이터타입
```

```python
x = (tidy.groupby('팀')['경기일자'].diff().dt.days == 1 ) #팀을 기준으로 경기일자를 하루 단위로 가져온다.(차집합)
```

```python
tidy['경기일자'][x]
```

```python
ab = (tidy.groupby('팀')['경기일자'].diff().dt.days -1 )
```

```python
tidy['휴식일'] = tidy.sort_values('경기일자').groupby('팀').경기일자.diff().dt.days - 1
#휴식일이라는 컬럼을 만들고 위의 예시들을 한줄로 요약해 만듬.
```
