```python
sat=pd.read_csv('2014_korea_sat_scores.csv',encoding='euc-kr')
sat.head()
```

```python
sat.shape
```

```python
sat['도수(명)']
```

```python
(sat['도수(명)']<20000).head()
```

```python
(sat['도수(명)']<20000)&(sat['비율(%)']) #데이터프레임에서 뽑아낸 시리즈 객체끼리의 비교
```

```python
l_sat=(sat['도수(명)']<20000)&(sat['비율(%)']) #데이터프레임에서 뽑아낸 시리즈 객체끼리의 비교
```

```python
sat_logic=sat[l_sat]#마스킹 기법. 마스킹해서 데이터 프레임에 넣어 시리즈들을 다시 데이터 프레임 객체로 만든다.
sat_logic
```
