```python
#데이터 합치기(접합 과 병합)
```

```python
import pandas as pd
import numpy as np

#접합부터


columns = np.array(['봄','여름','가을','겨울'])
```

```python
data = np.array([['A0','A1','A2','A3'],
                 ['B0','B1','B2','B3'],
                 ['C0','C1','C2','C3'],
                 ['D0','D1','D2','D3']])
```

```python
df1 = pd.DataFrame(data.T,columns=columns) #T를 붙이면 반전되어 열단위로 들어간다. 참고로 array의 함수니 numpy 전용이다.
```

```python
data1 = np.array([['A4','A5','A6','A7'],
                 ['B4','B5','B6','B7'],
                 ['C4','C5','C6','C7'],
                 ['D4','D5','D6','D7']])
```

```python
df2 = pd.DataFrame(data1.T,columns=columns)
```

```python
df_con = pd.concat([df1,df2]) #디폴트는 행단위 . 열의 이름은 같아야한다,.
```

```python
df_con
```

```python
index_r = list(range(0,8))
```

```python
df_con.index = index_r
```

```python
df_con
```

```python
df_con1 = pd.concat([df1,df2],ignore_index=True)
```

```python
df_con1
```

```python
df_h = pd.concat([df1,df2],axis=1)
```

```python
#데이터를 join
```

```python
data3 = np.array([['A8','A9','A10','A11'],
                 ['B8','B9','B10','B11'],
                 ['E8','E9','E10','E11'],
                 ['F8','F9','F10','F11']])
```

```python
columns3 = np.array(['봄','여름','춘분','추분'])
```

```python
df3 = pd.DataFrame(data3.T,columns=columns3)
```

```python
df3
```

```python
df4 = pd.concat([df1,df3],join='inner',ignore_index=True)
```

```python
df4 #뒤쪽 컬럼이 달라서 제외되어버린다.
```

```python
df5 = pd.concat([df1,df3],axis=1, join='inner') #열단위로 하니 행이 맞아서
```

```python
df5
```

```python
df6 = pd.concat([df1,df3],axis=1, join='outer')
```

```python
df6
```

```python
pd.concat([df1,df3],axis=1, keys=['사계절'],join='inner') #컬럼 위에 사계절이 뜬다.
```

```python
pd.concat([df1,df3],axis=1, keys=['사계절','춘추분'],join='inner')
```
