```python
import pandas as pd
import numpy as np
#키와 값의 개수가 같아야한다.
#중복된 키를 기준으로 행을 늘려나간다.
df1 = pd.DataFrame({'key':list('bbacaab'),'data1':range(7)})
df2 = pd.DataFrame({'key':list('abd'),'data2':range(3)})
df3 = pd.DataFrame({'key':list('bbacaab'),'data3':range(7)})
df4 = pd.DataFrame({'key':list('ababd'),'data4':range(5)})
df5 = pd.DataFrame({'lkey':list('ababd'),'data5':range(5)})
df6 = pd.DataFrame({'rkey':list('abd'),'data6':range(3)})
pd.merge(df3,df4, on='key', how='inner')
```

```python
pd.merge(df1,df2,on="key") #inner가 defualt
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right") #df2를 기준으로 df1을 갖다 붙인다.
```

```python
pd.merge(df5,df6,left_on='lkey',right_on='rkey')
```

```python
korea_2012 = pd.read_excel('korea_pop_2012.xls',encoding ='cp494')
korea_2012.index
```

```python
korea_2012.columns = ['','2008','2009','2010','2011','2012']
korea_2012.index = korea_2012['']
korea_2012 = korea_2012.drop([''],axis=1)
korea_2012                   
```

```python
korea_2017 = pd.read_excel('korea_pop_2017.xls',encoding='cp494')
korea_2017.index
korea_2017.columns = ['','2013','2014','2015','2016','2017']
korea_2017.index = korea_2017['']
korea_2017 = korea_2017.drop([''],axis=1)
korea_2017                   
```

```python
korea_2017.index==korea_2012.index
```

```python
korean_pop=pd.merge(korea_2012,korea_2017,left_index=True,right_index=True)
```

```python
korean_pop.get_dtype_counts()
```

```python
for i in range(0,18) :
    a = korea_pop.iloc[i].str.replace(',','')
    korea_pop.iloc[i] = a
```

```python
korea_pop.head()
```

```python
korea_pop.get_dtype_counts()
```

```python
korea_pop.isnull().sum()
```

```python
for i in range(0,18) :
    a = korea_pop.iloc[i].str.replace('-','NaN')
    korea_pop.iloc[i] = a
```

```python
korea_pop = korea_pop.astype('float64')
```

```python
korea_pop.isnull().sum()
```

```python
korea_pop = korea_pop.fillna(0)
```

```python
korea_pop = korea_pop.astype('int64')
```

```python
korea_pop = korea_pop.drop(["계"],axis=0)
```
