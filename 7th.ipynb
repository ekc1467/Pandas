```python
cat_1 = pd.Categorical(['a','c','c',np.nan],categories=['b','a','c']) #b,a,c를 제외한 나머지는 넣을 수 없음.
```

```python
cat_1 #b,a,c를 제외한 나머지는 넣을 수 없음.
```

```python
cat_1.dtype
```

```python
df = pd.DataFrame({'cat':cat_1, 's':['a','c','c',np.nan]}) #서로 다른 타입을 대입
```

```python
df  #서로 다른 타입의 컬럼 2줄이 만들어짐
```

```python
df.select_dtypes(exclude=['object'])
```
