```python
movies = pd.read_csv('movies.csv', usecols=['Film','Year'])
movies

```

```python
movies_add=pd.read_csv('movies.csv')
movies_add.shape
movies['has_seen']=0 #컬럼을 추가한다.
movies
```

```python
movies_add.columns
```

```python
movies['has_seen'] = (movies_add['Audience score %']+movies_add['Rotten Tomatoes %'])//2 #새로 만든 컬럼에 대입한다.
```

```python
movies
```

```python
movies['has_seen'].isnull().sum() #NaN(누락값)과의 연산결과는 NaN만 나온다. 항상 누락값이 있는 지 확인해보아야 한다.
```

```python
movies_add['Worldwide Gross'].dtype #모든 연산을 해볼 때 타입을 확인 해보는 것이 좋다.
```

```python
d=movies_add['Worldwide Gross'].str.replace('$','').astype('float64') #$를 떼어내고 float64로 타입을 바꿔라.
```

```python
movies.columns.get_loc('has_seen') #컬럼의 위치
```

```python
movies.insert(movies.columns.get_loc('has_seen')+1,"원화환산",d*1200) #제로부터 시작하니 +1을 해서 원하는 장소에 데이터를 인설트.
```
